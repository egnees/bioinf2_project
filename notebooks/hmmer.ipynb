## Подготавливаем hmmer (скачиваем, индексируем):


```python
%%bash
wget http://eddylab.org/software/hmmer/hmmer-3.3.2.tar.gz
tar xf hmmer-3.3.2.tar.gz
cd hmmer-3.3.2
```

```python
%%bash
cd hmmer-3.3.2
./configure
make
make install
```
